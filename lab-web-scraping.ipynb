```python
"""
1) Import Libraries
Load tools needed for scraping and data handling
"""
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.parse import urljoin


"""
2) Define Function
Create reusable function with filtering conditions
"""
def scrape_books(min_rating, max_price):

    """
    3) Set Website URL
    Define starting point of scraping
    """
    base_url = "https://books.toscrape.com/"
    current_url = urljoin(base_url, "catalogue/page-1.html")

    """
    4) Prepare Rating Conversion
    Convert rating text into numeric values
    """
    rating_map = {
        "One": 1,
        "Two": 2,
        "Three": 3,
        "Four": 4,
        "Five": 5
    }

    """
    5) Initialize Data Storage
    Store extracted book data
    """
    books_data = []

    """
    6) Loop Through Pages
    Iterate through all pages
    """
    while current_url:

        response = requests.get(current_url)
        soup = BeautifulSoup(response.text, "html.parser")

        """
        7) Extract Books List
        Get all books from current page
        """
        books = soup.select("article.product_pod")

        for book in books:

            """
            8) Collect Basic Book Data
            Extract title, price, rating, availability
            """
            title = book.h3.a["title"]

            price_text = book.select_one("p.price_color").text.strip()
            price = float(price_text.replace("£", "").replace("Â", ""))

            rating_class = book.select_one("p.star-rating")["class"]
            rating_word = rating_class[1]
            rating = rating_map[rating_word]

            availability = book.select_one("p.instock.availability").text.strip()

            """
            9) Access Book Details Page
            Build full URL for each book
            """
            relative_link = book.h3.a["href"]
            book_url = urljoin(current_url, relative_link)

            """
            10) Apply Filters
            Select books matching conditions
            """
            if rating >= min_rating and price <= max_price:

                """
                11) Extract Detailed Information
                Get extra data from book page
                """
                detail_response = requests.get(book_url)
                detail_soup = BeautifulSoup(detail_response.text, "html.parser")

                table_rows = detail_soup.select("table.table.table-striped tr")
                product_info = {}

                for row in table_rows:
                    key = row.th.text.strip()
                    value = row.td.text.strip()
                    product_info[key] = value

                upc = product_info.get("UPC", "")

                breadcrumb_items = detail_soup.select("ul.breadcrumb li a")
                genre = breadcrumb_items[2].text.strip() if len(breadcrumb_items) > 2 else ""

                description_tag = detail_soup.select_one("#product_description")
                if description_tag:
                    description = description_tag.find_next("p").text.strip()
                else:
                    description = ""

                """
                12) Store Book Data
                Save extracted book info
                """
                books_data.append({
                    "UPC": upc,
                    "Title": title,
                    "Price (£)": price,
                    "Rating": rating,
                    "Genre": genre,
                    "Availability": availability,
                    "Description": description
                })

        """
        13) Move to Next Page
        Continue until last page
        """
        next_button = soup.select_one("li.next a")
        if next_button:
            next_page = next_button["href"]
            current_url = urljoin(current_url, next_page)
        else:
            current_url = None

    """
    14) Create DataFrame
    Convert data into table format
    """
    df = pd.DataFrame(books_data)

    """
    15) Return Results
    Output final dataset
    """
    return df


"""
16) Run the Function
Execute with given conditions
"""
df = scrape_books(4, 20)

print(df.head())
print(f"\nTotal books found: {len(df)}")
```
